```python
from matplotlib import image 
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
```

```python
data = image.imread('D:\sem-5-practicals\sem-5\deep-rl-lab\opera_house.jpg')
print(data.dtype)
print(data.shape)
pyplot.imshow(data)
pyplot.show()
```

```python
#To Resize the Image
img1 = Image.open('D:\sem-5-practicals\sem-5\deep-rl-lab\opera_house.jpg')
new_image = img1.resize((500,500))
new_image.save('myimage_500.jpg')
pyplot.imshow(new_image)
```

```python
# To convert image to numpy array
img2= Image.open('D:\sem-5-practicals\sem-5\deep-rl-lab\opera_house.jpg')
data2 = asarray(img2)
print(data2.shape)
image2 = Image.fromarray(data2)
print(image2.format)
print(image2.mode)
print(image2.size)
```

```python
# To flip the image
hoz_flip = img2.transpose(Image.FLIP_LEFT_RIGHT)
ver_flip = img2.transpose(Image.FLIP_TOP_BOTTOM)
pyplot.subplot(3,1,1)
pyplot.imshow(img2)
pyplot.subplot(3,1,2)
pyplot.imshow(hoz_flip)
pyplot.subplot(3,1,3)
pyplot.imshow(ver_flip)
pyplot.show()
```

```python
pyplot.subplot(4,1,1)
pyplot.imshow(img2)
pyplot.subplot(4,1,2)
pyplot.imshow(img2.rotate(45))
pyplot.subplot(4,1,3)
pyplot.imshow(img2.rotate(90))
pyplot.subplot(4,1,4)
pyplot.imshow(img2.rotate(270))
pyplot.show()
```
